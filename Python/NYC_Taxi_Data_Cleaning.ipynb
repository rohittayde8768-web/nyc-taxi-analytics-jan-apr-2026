```python
import pandas as pd

# Numeric columns
numeric_cols = [
    "VendorID",
    "passenger_count",
    "trip_distance",
    "RatecodeID",
    "PULocationID",
    "DOLocationID",
    "payment_type",
    "fare_amount",
    "extra",
    "mta_tax",
    "tip_amount",
    "tolls_amount",
    "improvement_surcharge",
    "total_amount",
    "congestion_surcharge",
    "Airport_fee",
    "cbd_congestion_fee",
    "trip_duration_minutes"
]

# String columns
string_cols = [
    "store_and_fwd_flag",
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]

# Clean numeric columns
for col in numeric_cols:
    test[col] = pd.to_numeric(test[col], errors="coerce").fillna(0)

# Clean string columns
for col in string_cols:
    test[col] = (
        test[col]
        .fillna("")      # Replace NULL with empty string
        .astype(str)     # Convert to string
        .str.strip()     # Remove leading/trailing spaces
    )
```
