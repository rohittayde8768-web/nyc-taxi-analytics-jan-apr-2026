```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_parquet("nyc_taxi_cleaned_final.parquet")

df.shape
```

```python
df['pickup_date'] = df['tpep_pickup_datetime'].dt.date

df['pickup_month'] = df['tpep_pickup_datetime'].dt.month_name()

df['pickup_day'] = df['tpep_pickup_datetime'].dt.day_name()

df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour

df['pickup_weekday'] = df['tpep_pickup_datetime'].dt.dayofweek
```

```python
df['pickup_hour'].value_counts().sort_index().plot(
    kind='bar',figsize=(10,5))

plt.title("Trips by Pickup Hour")
plt.xlabel("Hour")
plt.ylabel("Number of Trips")

plt.show()
```

```python
df['payment_type'].value_counts().sort_index().plot(
    kind='bar',
    figsize=(6,4)
)

plt.title("Payment Type Distribution")
plt.xlabel("Payment Type")
plt.ylabel("Trips")

plt.show()
```

```python
df['VendorID'].value_counts().sort_index().plot(
    kind='bar',
    figsize=(5,4)
)

plt.title("Trips by Vendor")
plt.xlabel("Vendor")
plt.ylabel("Trips")

plt.show()
```

```python
df['PULocationID'].value_counts().head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Top 10 Pickup Locations")
plt.xlabel("Pickup Location ID")
plt.ylabel("Trips")

plt.show()
```

```python
df['DOLocationID'].value_counts().head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Top 10 Drop-off Locations")
plt.xlabel("Drop-off Location ID")
plt.ylabel("Trips")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['trip_distance'], bins=50)

plt.title("Trip Distance Distribution")
plt.xlabel("Distance (Miles)")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['trip_duration_minutes'], bins=50)

plt.title("Trip Duration Distribution")
plt.xlabel("Minutes")
plt.ylabel("Frequency")

plt.show()
```

```python
df.to_parquet("nyc_taxi_final.parquet", index=False)
```
